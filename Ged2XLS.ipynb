```python
# FIRST LINE

###############################################
# CHANGE THE ONE LINE IN THIS SECTION
###############################################
input_file = '/content/sample_data/The English and British Kings and Queens.ged'
###############################################

###############################################
# NO CHANGES PAST THIS POINT
###############################################
import os
import xlsxwriter

from gedcom.parser import Parser
from gedcom.element.individual import IndividualElement

name_part = os.path.splitext(input_file)[0]
xls_file_path = name_part + '.xlsx'

###############################################
# create the spreadsheet file
###############################################
workbook = xlsxwriter.Workbook(xls_file_path)
worksheet = workbook.add_worksheet()

###############################################
# Initialize the parser
###############################################
gedcom_parser = Parser()

gedcom_parser.parse_file(input_file, False)  # Disable strict parsing

###############################################
# Work through GEDCOM file and grab person data
###############################################
root_child_elements = gedcom_parser.get_root_child_elements()

person_count = 0

for element in root_child_elements:
    if isinstance(element, IndividualElement):

        # process individual
        first_name = last_name = pointer = ''
        birth_date = birth_place = death_date = death_place = ''
        father_first_name = father_last_name = father_pointer = ''
        mother_first_name = mother_last_name = mother_pointer = ''

        try:
            first_name, last_name = element.get_name()
            gender = element.get_gender()
            birth_date, birth_place, birth_sources = element.get_birth_data()
            death_date, death_place, death_sources = element.get_death_data()
            pointer = element.get_pointer()

        except Exception as e:
            print("ERROR:", str(e))

        # process parents
        try:
            parents = gedcom_parser.get_parents(element)
            for parent in parents:
                if parent.get_gender() == 'M':
                    father_first_name, father_last_name = parent.get_name()
                    father_pointer = parent.get_pointer()
                else:
                    mother_first_name, mother_last_name = parent.get_name()
                    mother_pointer = parent.get_pointer()
        except Exception as e:
            print("ERROR:", str(e))

        ##########################################
        # write to spreadsheet
        ##########################################
        worksheet.write(person_count, 0, first_name)
        worksheet.write(person_count, 1, last_name)

        worksheet.write(person_count, 2, birth_date)
        worksheet.write(person_count, 3, birth_place)

        worksheet.write(person_count, 4, death_date)
        worksheet.write(person_count, 5, death_place)

        worksheet.write(person_count, 6, father_first_name)
        worksheet.write(person_count, 7, father_last_name)

        worksheet.write(person_count, 8, mother_first_name)
        worksheet.write(person_count, 9, mother_last_name)

        worksheet.write(person_count, 10, pointer)
        worksheet.write(person_count, 11, father_pointer)
        worksheet.write(person_count, 12, mother_pointer)

        # show some data
        print(first_name, last_name)

        # next person
        person_count += 1

workbook.close()

# LAST LINE
```

```python
import sys
from ged4py.parser import GedcomReader

with GedcomReader(sys.argv[1]) as parser:
    # iterate over each FAM record in a file
    for i, fam in enumerate(parser.records0("FAM")):

        print(f"{i}:")

        # Get records for spouses, FAM record contains pointers to INDI
        # records but sub_tag knows how to follow the pointers and return
        # the referenced records instead.
        husband, wife = fam.sub_tag("HUSB"), fam.sub_tag("WIFE")
        if husband:
            print(f"    husband: {husband.name.format()}")
        if wife:
            print(f"    wife: {wife.name.format()}")

        # Get _value_ of the MARR/DATE tag
        marr_date = fam.sub_tag_value("MARR/DATE")
        if marr_date:
            print(f"    marriage date: {marr_date}")

        # access all CHIL records, sub_tags method returns list (possibly empty)
        children = fam.sub_tags("CHIL")
        for child in children:
            # print name and date of birth
            print(f"    child: {child.name.format()}")
            birth_date = child.sub_tag_value("BIRT/DATE")
            if birth_date:
                print(f"        birth date: {birth_date}")
```
